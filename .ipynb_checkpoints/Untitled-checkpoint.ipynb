```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("NewlyAddedTestsReport.csv")

df.dtypes

df.head(10)

X = np.array(df.Component.unique())
# print X

y = np.unique(df.Component, return_counts=True)
Y=y[1]
orders = y[0]

size = len(df.Component.unique())

# sns.set(style="whitegrid")

# print y
```

```python
fig = plt.figure(figsize=(size+3,size+3))

sns.set_style("whitegrid")

colors = sns.color_palette("cubehelix", len(df.Component.unique()) + 5)

for i in range(np.shape(X)[0]):
    g = sns.barplot(x=X[i:i+1], y=Y[i:i+1], color=colors[i], order = X, url = X[i])
    g.text(i, Y[i] + 0.5, Y[i], color='black', ha="center", weight="bold")


sns.despine(left=True)
plt.savefig("test1.svg")
```
