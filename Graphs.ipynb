```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("NewlyAddedTestsReport.csv")
```

```python
X, Y = np.unique(df.Component, return_counts=True)

size = np.shape(X)[0]
```

```python
fig = plt.figure(figsize=(size+3,size+3))
sns.set_style("whitegrid")

colors = sns.color_palette("cubehelix", len(df.Component.unique()) + 5)

for i in range(size):
    g = sns.barplot(x=X[i:i+1], y=Y[i:i+1], color=colors[i], order = X, url = X[i]+'.svg')
    g.text(i, Y[i] + 0.5, Y[i], color='black', ha="center", weight="bold")


sns.despine(left=True)
plt.savefig("test1.svg")
```
